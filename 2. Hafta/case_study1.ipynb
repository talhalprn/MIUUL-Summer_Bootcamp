```python
# görev1 : Seaborn kütüphanesi içerisinden Titanicveri setini tanımlayınız. 

import seaborn as sns
df = sns.load_dataset("titanic")
```

```python
# görev2 : Titanic veri setindeki kadın ve erkek yolcuların sayısını bulunuz.

print(df["sex"].value_counts())
```

```python
#görev3 : Her bir sutuna ait unique değerlerin sayısını bulunuz.

df.nunique()
```

```python
#görev4 : pclass değişkenini nunique değerlerinin sayısını bulunuz.

df["pclass"].nunique()
```

```python
#görev5 : pclass ve parch değişkenlerini nunique değerlerinin sayısını bulunuz

df[["pclass", "parch"]].nunique()
```

```python
#görev6 : embarked değişkeninin tipini kontrol ediniz. Tipini category olarak değiştiriniz ve tekrardan kontrol ediniz

print(df["embarked"].dtype)

df["embarked"] = df["embarked"].astype("category")

print(df["embarked"].dtype)
```

```python
#görev7 : embarked değeri "C" olanların değerini gösteriniz

df.loc[df["embarked"] == "C"]
```

```python
#Görev8 : embarked değeri s olmayanların bütün bilgilerini gösteriniz.

df.loc[df["embarked"] != "S"]
```

```python
#Görev 9: Yaşı 30 dan küçük ve kadın olan yolcuların tüm bilgilerini gösteriniz.

df.loc[ (df["age"] < 30) & (df["sex"] == "female")]
```

```python
#Görev 10: Fare'i 500'den büyük veya yaşı 70 den büyük yolcuların bilgilerini gösteriniz.

df.loc[ (df["fare"] < 500) | (df["age"] > 70) ]
```

```python
#Görev11 : Her bir değişkendeki boş değerlerin toplamını bulunuz.

df.isnull().sum()
```

```python
#Görev 12: who değişkenini dataframe’den çıkarınız.

df_2 = df.drop("who", axis = 1)

df_2.head()
```

```python
#Görev13 : deck değikenindeki boş değerleri deck değişkenin en çok tekrar eden değeri (mode) ile doldurunuz.

df["deck"].fillna( df["deck"].mode().values[0], inplace = True )

df["deck"].isnull().sum()
```

```python
mode = df["deck"].mode().values[0]
print(mode)
```

```python
df.loc[df["deck"].isna() , 'deck'] = mode
```

```python
df["deck"].unique()
```

```python
# Görev 14: age değişkenindeki boş değerleri age değişkenin medyanı ile doldurunuz.

median = df["age"].median()
print(median)
```

```python
df.loc[df["age"].isna() , 'age'] = median
df["age"].isnull().sum()
```

```python
#Görev15 : survived değişkeninin pclass ve cinsiyet değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.

df.groupby(["pclass","sex"]).agg( {"survived": ["sum","count","mean"]})
```

```python
#Görev 16: 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazın. 
#Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz oluşturunuz.
#(apply ve lambda yapılarını kullanınız)

df["age_flag"] = df["age"].apply(lambda age: 1 if age < 30 else 0 )
df[["age","age_flag"]]
```

```python
#Görev 17 : Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.

import seaborn as sns

df_3 = sns.load_dataset("tips")
df_3.head()
```

```python
#Görev 18 : Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.

df_3.columns
```

```python
df_3["time"].unique()
```

```python
df_3.groupby("time").agg({"total_bill": ["min","max","mean"]})
```

```python
#Görev 19 : Günlere ve time göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.

df_3.groupby(["time","day"]).agg({"total_bill": ["min","max","mean"]})
```

```python
#Görev 20 : Lunch zamanına ve kadın müşterilere ait total_bill ve tip değerlerinin day'e göre toplamını, min, max ve ortalamasını bulunuz.

df_3.loc[(df_3.time == "Lunch") & (df_3.sex == "Female")].groupby("day").agg({"total_bill":["mean","sum","max","min"], "tip":["mean","sum","max","min"]})
```

```python
#Görev 21: size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir? (loc kullanınız)

df_3.loc[(df_3["size"] < 3) & (df_3["total_bill"] > 10),"total_bill"].mean()
```

```python
# Görev 22: total_bill_tip_sum adında yeni bir değişken oluşturunuz. Her bir müşterinin ödediği totalbill ve tip in toplamını versin.

df_3["total_bill_tip_sum"] = df_3["total_bill"] + df_3["tip"]

df_3.head()
```

```python
#Görev 23 : total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.

new_df = df_3.sort_values("total_bill_tip_sum", ascending = False).head(30)
new_df
```
